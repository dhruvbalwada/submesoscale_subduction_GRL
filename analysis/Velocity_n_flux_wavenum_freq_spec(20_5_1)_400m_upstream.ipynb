# Code to calculate the $\omega$ and $\kappa$ spectra of the vertical velocity and vertical flux of tracer. 
At base of ML for each simulation. And at a deeper level (400m).

This is just an example script - one could use the same steps to do this calculation anywyere else. Caution: the algorithm is not optimized and uses up a lot of memory for large datasets. 

```python
# load appropriate modules
import numpy as np
import xarray as xr
from xmitgcm import open_mdsdataset
from matplotlib import pyplot as plt
%matplotlib inline
from xgcm import Grid
from scipy import linalg as lin
from scipy import signal as sig
from scipy import fftpack as fft
from scipy import interpolate as naiso
from matplotlib import colors, ticker, cm
import xrft
#import gsw
```

```python
def azimuthal_avg(k, l, f, N, nfactor):
    """
    Takes the azimuthal average of a given field.
    """
    k = k.values; l = l.values
    kk, ll = np.meshgrid(k, l)
    K = np.sqrt(kk**2 + ll**2)
    nbins = int(N/nfactor)
    if k.max() > l.max():
        ki = np.linspace(0., l.max(), nbins)
    else:
        ki = np.linspace(0., k.max(), nbins)

    kidx = np.digitize(np.ravel(K), ki)
    area = np.bincount(kidx)

    kr = np.bincount(kidx, weights=K.ravel()) / area

    if f.ndim == 2:
        iso_f = np.ma.masked_invalid(np.bincount(kidx,
                                    weights=f.data.ravel())
                                    / area) * kr
    else:
        raise ValueError('The data has too many or few dimensions. '
                        'The input should only have the two dimensions '
                        'to take the azimuthal averaging over.')
    return kr, iso_f
```

```python
# Make dictionary
res_time = {"20":{'dt':1200, 'iter1': 5702400, 'iterend':5715360},
            "05":{'dt':300, 'iter1':10368000, 'iterend':10402560}, 
            "01":{'dt':40 , 'iter1':90, 'iterend':345600}} # this last of 1km simulation will be updated

for keys in res_time.keys():
    if keys == '01':
        res_time[keys]['dirname']='/swot/SUM04/dbalwada/channel_beta/{0}km/tracer_exp2/'.format(keys)
    else:
        res_time[keys]['dirname']='/swot/SUM05/dbalwada/channel_beta/{0}km/tracer_exp2/'.format(keys)
```

```python
res = sorted(res_time.keys())
```

```python
# pick up data in "=" (native) format

for key in res:
    npick = 1 # pick up every npick'th file 

    dt = int(3600/res_time[key]['dt'])
    T1 = res_time[key]['iter1']
    T2 = res_time[key]['iterend']
    
    T95 = T1 + 40*24*dt
    T110 = T1 + 160*24*dt
    
    iters = range(T95, T110+dt, dt*npick)

    ds = open_mdsdataset(res_time[key]['dirname'],  iters=iters,
                                 delta_t=res_time[key]['dt'], prefix=['W','PTRACER01']
                         ,ignore_unknown_vars=True, geometry='cartesian')
    
    res_time[key]['data'] = ds
```

```python
W_us = {}
C_us = {}

for key in res: 
    print(key)
    ds = res_time[key]['data']
    grid = Grid(ds, periodic='X')
    
    W_temp = ds.W.sel(XC=xsel, YC=ysel)
    W_int = grid.interp(W_temp, 'Z', boundary='extend')
    W_int.name = 'W'
    
    W_sel = W_int.sel(Z = zsel, method='nearest')
    
    C_sel = ds.PTRACER01.sel(XC=xsel, YC=ysel).sel(Z = zsel, method='nearest')
    
    W_us[key] = W_sel.load()
    C_us[key] = C_sel.load()
```

```python
# Calculate the W velocity power spectra (time)
# Calculate the WC cross spectra (time)
# this is done indepdendently at every grid point

Wps = {}
WCcs = {}

for key in res:
    print(key)
    Wps[key] = xrft.power_spectrum(W_us[key], dim=['XC','YC','time'], window='hanning')
    WCcs[key] = xrft.cross_spectrum(W_us[key], C_us[key],
                                    dim = ['XC','YC','time'], window='hanning')
```

```python
# Add the isotropic calculation (k,l) -> (k_r) (azimuthally average)
def azimuthal_avg(k, l, f, N, nfactor):
    """
    Takes the azimuthal average of a given field.
    """
    k = k.values; l = l.values
    kk, ll = np.meshgrid(k, l)
    K = np.sqrt(kk**2 + ll**2)
    nbins = int(N/nfactor)
    if k.max() > l.max():
        ki = np.linspace(0., l.max(), nbins)
    else:
        ki = np.linspace(0., k.max(), nbins)

    kidx = np.digitize(np.ravel(K), ki)
    area = np.bincount(kidx)

    kr = np.bincount(kidx, weights=K.ravel()) / area

    if f.ndim == 2:
        iso_f = np.ma.masked_invalid(np.bincount(kidx,
                                    weights=f.data.ravel())
                                    / area) * kr
    else:
        raise ValueError('The data has too many or few dimensions. '
                        'The input should only have the two dimensions '
                        'to take the azimuthal averaging over.')

    return kr, iso_f

def iso_spec_over_depth(ds):
    
    Zlen = len(ds.Z)
    klen = len(ds.freq_XC)
    omegalen = len(ds.freq_time)
    
    nfactor = 4. 
    ds_iso = np.zeros([omegalen , Zlen, int(klen/nfactor+1)])

    for j in range(0,omegalen):
        for i in range(0,Zlen):
            kr, ds_iso[j,i,:] = azimuthal_avg(ds.freq_XC, ds.freq_YC, ds.isel(Z=i, freq_time=j), 
                            len(ds.freq_XC), nfactor)
    kr[0] = 0.
    ds_iso_xr = xr.DataArray(ds_iso, coords={'freq_time':ds.freq_time.values, 'Z': ds.Z.values, 'kr':kr }, 
                             dims=['freq_time','Z', 'kr'])
    
    return ds_iso_xr
        
def iso_spec(ds):
    
    #Zlen = len(ds.Z)
    klen = len(ds.freq_XC)
    omegalen = len(ds.freq_time)
    
    nfactor = 4. 
    ds_iso = np.zeros([omegalen ,  int(klen/nfactor+1)])

    for j in range(0,omegalen):
        
            kr, ds_iso[j,:] = azimuthal_avg(ds.freq_XC, ds.freq_YC, ds.isel( freq_time=j), 
                            len(ds.freq_XC), nfactor)
    kr[0] = 0.
    ds_iso_xr = xr.DataArray(ds_iso, coords={'freq_time':ds.freq_time.values, 'kr':kr }, 
                             dims=['freq_time','kr'])
    
    return ds_iso_xr
```
